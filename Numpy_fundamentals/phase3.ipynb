## Advance operation with Business  examples

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
### Data Structure: [restaurant_id,2021,2022,2023,2024]

sales_data = np.array([
    [1,150000,180000,220000,250000], # Paradise Biryani
    [2,120000,140000,160000,190000], # Biryani House
    [3,200000,230000,260000,300000], # Pizza Hub
    [4,180000,210000,240000,270000], # Burger King
    [5,160000,185000,205000,230000], # Chai Point
    
])

print ("=== Sales Data ===")
print("\n Sales Data shape:",sales_data.shape)
print("\n Sample data for 1st 3 rest: \n",sales_data[:3])
```

```python
# total sales per year

print(np.sum(sales_data,axis= 0,))
yearly_total = np.sum(sales_data[:,1:],axis = 0)
print("\n Yearly Total Sales: ",yearly_total)

summ = np.sum(yearly_total)
print("\n Total Sales: ",summ)


#Minimum sales per restaraunt

min_sales = np.min(sales_data[:,1:],axis = 1)
print("\n Minimum sales per restaurant: ",min_sales)

# maximum sales per  restaurant

max_sales = np.max(sales_data[:,1:],axis = 0)
print("\n Maximum sales per year: ",max_sales)


# Average sales per restaurant

avg_sales = np.mean(sales_data[:,1:],axis = 1)
print("\n Average sales per restaurant:",avg_sales)

# Cummulative sales per restaurant
cum_sales = np.cumsum(sales_data[:,1:],axis = 1)
print("\n Cummulative sales per restaurant: ",cum_sales)
```

```python
plt.figure(figsize = (12,6))
plt.plot(np.mean(cum_sales,axis = 0))
plt.title("Cummulative sales per restaurant")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print("\n Addition of two vectors:",vector1 + vector2)

print("\n Subtraction of two vectors:",vector1 - vector2)
print("\n Multiplication of two vectors:",vector1 * vector2)
print("\n Division of two vectors:",vector1 / vector2)
print("\n Exponentiation of two vectors:",vector1 ** vector2)
print("\n Dot Product of two vectors:",np.dot(vector1,vector2))
```

```python
angle = np.arccos(np.dot(vector1,vector2) / (np.linalg.norm(vector1)* np.linalg.norm(vector2)))
print("\n Angle between two vectors in radians:",angle)
```
